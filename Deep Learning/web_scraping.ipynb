```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL="https://www.values.com/inspirational-quotes"
r=requests.get(URL)
```

```python
soup=BeautifulSoup(r.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
quote={"Theme":[],"URL":[],"IMG":[],"Lines":[]}
```

```python
table=soup.find('div',attrs={'id':"all_quotes"})
```

```python
for row in table.findAll("div",attrs={'class':'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}):
    print(row)
    quote['Theme'].append(row.h5.text)
    quote['URL'].append(row.a['href'])
    quote['IMG'].append(row.img['src'])
    quote['Lines'].append(row.img['alt'].split("#")[0])
```

```python
df=pd.DataFrame(quote)
```

```python
df.head()
```

```python
df.to_csv("df.csv")
```
