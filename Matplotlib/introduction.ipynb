```python
blood_sugar=[100,234,764,125,80,80,30,40,678,875,893]
plt.hist(blood_sugar,rwidth=0.5,bins=5) #rwidth is width of the barhistogram
plt.show()
```

```python
blood_sugar_men=[80,85,90,95,100,105,110,115,120,125,130,135,140,145,150]
blood_sugar_women=[83,86,93,96,103,106,113,116,123,126,133,136,143,146,153]
plt.hist([blood_sugar_men,blood_sugar_women],rwidth=0.9,color=['green','blue'],bins=[80,100,125,155],label=["blood_sugar men","blood_sugar women"])
plt.xlabel("Sugar Level",size=20,color='r')
plt.ylabel("Number of patients",size=20,color='r')
plt.title("Blood Sugar Chart",size=30,color='navy')
plt.legend(loc='upper right',shadow=True)
plt.show()
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.5,histtype='step',color='pink')  # step,stepfilled,barstacked,bar..
plt.show()
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.5,histtype='stepfilled',color='pink')  # step,stepfilled,barstacked,bar..
plt.show()
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.5,histtype='barstacked',color='pink')  # step,stepfilled,barstacked,bar..
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
timesData = pd.read_csv('./cwurData.csv')
```

```python
timesData.tail()
```

```python
df = timesData.iloc[:100,:]
```

```python
# Creating trace1
trace1 = go.Scatter(x = df.world_rank,y = df.citations,mode = "lines",name = "citations",marker = dict(color = 'rgba(16, 112, 2, 0.8)'),text= df.institution)

# Creating trace2
trace2 = go.Scatter(x = df.world_rank,y = df.quality_of_faculty,mode = "lines+markers",name = "teaching",marker = dict(color = 'rgba(80, 26, 80, 0.8)'),text= df.institution)
data = [trace1, trace2]

data = [trace1, trace2]
layout = dict(title = 'Citation and Teaching vs World Rank of Top 100 Universities',xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False))

fig = dict(data = data, layout = layout)
iplot(fig)
```
