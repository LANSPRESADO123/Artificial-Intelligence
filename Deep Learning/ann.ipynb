```python
x=pima.drop(columns=['Outcome'])
y=pima['Outcome']
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(x,y)
```

```python
a=pd.Series(model.feature_importances_,x.columns)
print(a.nlargest(9))
```

```python
a.nlargest(9).plot(kind='barh',color='Purple')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

```

```python
scaler = StandardScaler()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.fit_transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
y_train.shape
```

```python
nn_mlp = MLPClassifier(hidden_layer_sizes=(7,3,5),activation='relu',solver='lbfgs',random_state=1,max_iter=300)
nn_mlp.fit(x_train_std,y_train)
```

```python
nn_mlp.score(x_test_std,y_test)
```
