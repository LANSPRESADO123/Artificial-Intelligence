```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
iris=load_iris()

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
iris
```

```python
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df
```

- Dropping Irrelevant features

```python
df=df.drop(df[['petal length (cm)','petal width (cm)']],axis=1)
df
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
plt.xlabel('Sepal_Length')
plt.ylabel('Sepal_Width')
```

```python
km=KMeans(n_clusters=2)
y_predicted=km.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
y_predicted
```

- Clustering Petals

```python
df['cluster']=y_predicted
```

```python
irdf1 = df[df['cluster']==0]
irdf2 = df[df['cluster']==1]


plt.scatter(irdf1['sepal length (cm)'],irdf1['sepal width (cm)'])
plt.scatter(irdf2['sepal length (cm)'],irdf2['sepal width (cm)'])


plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'red',marker='*',label='centroid')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend();
```

```python
km.cluster_centers_
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['sepal length (cm)','sepal width (cm)']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Errors')
plt.plot(k_rng,sse)
```

- With 3 clusters

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['sepal length (cm)','sepal width (cm)']])

df['cluster']=y_predicted
irdf1 = df[df['cluster']==0]
irdf2 = df[df['cluster']==1]
irdf3 = df[df['cluster']==2]


plt.scatter(irdf1['sepal length (cm)'],irdf1['sepal width (cm)'])
plt.scatter(irdf2['sepal length (cm)'],irdf2['sepal width (cm)'])
plt.scatter(irdf3['sepal length (cm)'],irdf3['sepal width (cm)'])


plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'red',marker='*',label='centroid')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend();
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['sepal length (cm)','sepal width (cm)']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Errors')
plt.plot(k_rng,sse)
```
