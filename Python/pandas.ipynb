```python
import pandas as pd

flower_boxes = {'Color': ['Lavender','Lavender','Lavender','Blue','Blue','Red','Red','Red'],
         'Flower': ['Lavender','Lavender','Tulip','Lavender','Tulip','Tulip','Tulip','Lavender'],
         'Price': [10,15,5,5,10,15,15,5]
        }

df = pd.DataFrame(flower_boxes, columns= ['Color','Flower','Price'])
print (df)
```

```python
sel_color=df.loc[df['Color']=='Lavender']
print(sel_color)
```

```python
sel_color_and_shape=df.loc[(df['Color']=='Red') & (df['Flower']=='Lavender')]
print(sel_color_and_shape)
```

```python
sel_color_and_shape=df.loc[(df['Color']=='Red') | (df['Flower']=='Lavender')]
print(sel_color_and_shape)
```
